```python
from src.utils.lr_scheduler import CosineAnnealingWarmupRestarts
```

```python
import torch
import torch.nn as nn
x = nn.Linear(10,10)
optim = torch.optim.Adam(x.parameters(),lr=1e-4)
```

```python
c = CosineAnnealingWarmupRestarts(optim,
                                  first_cycle_steps=403,
                                  cycle_mult=1,
                                  max_lr=5e-4,
                                  min_lr=1e-5,
                                  warmup_steps=40,
                                  gamma=0.975)
```

```python
import matplotlib.pyplot as plt
lst = []
for i in range(403*150):
    c.step()
    lst.append(c.get_lr())
```

```python
plt.figure(figsize=(20,8))
plt.plot(lst)
```
